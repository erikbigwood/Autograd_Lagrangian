```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x*np.sin(x) - np.cos(x)

xgrid = np.linspace(-10,10,1000)

plt.plot(xgrid, f(xgrid))
plt.plot(xgrid, 0*xgrid)
```
